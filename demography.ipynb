```python
# -*- coding: utf-8 -*-

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


excel_file_path = 'C:\\Users\\16690\\Desktop\\问卷数据.xlsx'

# 读取 Excel 文件中名为 'Sheet1' 的工作表
df = pd.read_excel(excel_file_path, sheet_name='Sheet1')


# 打印 DataFrame 的前几行
print(df.head())
```

```python
# 提取问题10到问题17的数据
questions_10_to_17 = df.iloc[:, 15:49]
questions_10_to_17 = questions_10_to_17.astype(int)

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文显示
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 计算平均值
mean_values = questions_10_to_17.mean()

# 计算中位数
median_values = questions_10_to_17.median()

# 计算众数
mode_values = questions_10_to_17.mode().iloc[0]

# 计算标准差
std_deviation = questions_10_to_17.std()

print("\n平均值:")
print(mean_values)

print("\n中位数:")
print(median_values)

print("\n标准差:")
print(std_deviation)

print("\n众数:")
print(mode_values)
```

```python
import pandas as pd
import pingouin as pg

# 使用 alpha 函数计算 Cronbach's Alpha
alpha_result = pg.cronbach_alpha(data=questions_10_to_17)
cronbach =[pg.cronbach_alpha(data=questions_10_to_17.iloc[:, 0:7]),
           pg.cronbach_alpha(data=questions_10_to_17.iloc[:, 8:16]),
           pg.cronbach_alpha(data=questions_10_to_17.iloc[:, 17:24]),
           pg.cronbach_alpha(data=questions_10_to_17.iloc[:, 25:32])]

# 输出结果
print(alpha_result,'\n',cronbach[0],'\n',cronbach[1],'\n',cronbach[2],'\n',cronbach[3])
```

```python
import matplotlib.pyplot as plt

# 计算每个维度的平均值
learning_adaptation_means = df.filter(like='学习适应').mean()
life_adaptation_means = df.filter(like='生活适应').mean()
psychological_adaptation_means = df.filter(like='心理适应').mean()
cultural_adaptation_means = df.filter(like='文化适应').mean()

# 将平均值组合成一个新的DataFrame
means_df = pd.DataFrame({
    '学习适应': learning_adaptation_means,
    '生活适应': life_adaptation_means,
    '心理适应': psychological_adaptation_means,
    '文化适应': cultural_adaptation_means
})

# 可视化每个维度的平均值
means_df.plot(kind='bar', rot=0, title='各维度平均值')
plt.xlabel('维度')
plt.ylabel('平均值')
plt.show()
```

```python
# 计算每个维度的统计量
learning_adaptation_stats = df.filter(like='学习适应').describe()
life_adaptation_stats = df.filter(like='生活适应').describe()
psychological_adaptation_stats = df.filter(like='心理适应').describe()
cultural_adaptation_stats = df.filter(like='文化适应').describe()

# 打印每个维度的统计量
print("学习适应统计量：")
print(learning_adaptation_stats)

print("\n生活适应统计量：")
print(life_adaptation_stats)

print("\n心理适应统计量：")
print(psychological_adaptation_stats)

print("\n文化适应统计量：")
print(cultural_adaptation_stats)
```

```python
# 使用seaborn绘制箱线图
sns.boxplot(data=questions_10_to_17)
# 将xtick标签旋转90度
plt.xticks(rotation=90)
plt.title('Likert Scale Distribution')
plt.show()
```

```python
# 计算各个问题之间的相关性矩阵
correlation_matrix = questions_10_to_17.corr()

# 使用seaborn绘制相关性热力图
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
