```python
# Original configurations and scores
configurations = ['next_line-drrip', 'next_line-lru', 'next_line-ship', 'next_line-srrip', 'ip_stride-drrip', 'ip_stride-lru', 'ip_stride-ship', 'ip_stride-srrip']
average_scores = [0.888380, 0.885845, 0.875445, 0.882245, 0.858835, 0.818590, 0.891540, 0.839015]

# Separate configurations into groups
grouped_configurations = {'drrip': [], 'lru': [], 'ship': [], 'srrip': []}

for config, score in zip(configurations, average_scores):
    if 'drrip' in config:
        grouped_configurations['drrip'].append(score)
    elif 'lru' in config:
        grouped_configurations['lru'].append(score)
    elif 'ship' in config:
        grouped_configurations['ship'].append(score)
    elif 'srrip' in config:
        grouped_configurations['srrip'].append(score)

# Plotting
labels = ['drrip', 'lru', 'ship', 'srrip']
data = [grouped_configurations[label] for label in labels]

plt.figure(figsize=(10, 6))
plt.boxplot(data, labels=labels, showmeans=True, meanline=True)
plt.title('Average Scores for Different Configurations')
plt.xlabel('Configurations')
plt.ylabel('Average Score')
plt.ylim(0.8, 0.9)  # Set y-axis range for better visualization
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

configurations = ['drrip', 'lru', 'ship', 'srrip']
prefetchers = ['next_line-next_line-next_line', 'no-ip_stride-no']
average_scores = [
    [0.888380, 0.858835],  # drrip configurations
    [0.885845, 0.818590],  # lru configurations
    [0.875445, 0.891540],  # ship configurations
    [0.882245, 0.839015]   # srrip configurations
]

bar_width = 0.35
index = np.arange(len(configurations))

plt.figure(figsize=(12, 6))

for i in range(2):
    plt.bar(index + i * bar_width, [score[i] for score in average_scores], width=bar_width, label=f'Prefetcher {i + 1}')

plt.title('Average Scores for Different Replacement Policies and Prefetchers')
plt.xlabel('Replacement Policy')
plt.ylabel('Average Score')
plt.xticks(index + bar_width / 2, configurations)
plt.ylim(0.8, 0.9)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), title='Prefetcher')
plt.show()

```
